```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('/content/sample_data/Iris.csv')
```

```python
dataset
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(dataset.drop('Species',axis=1))
```

```python
scaled_features=scaler.transform(dataset.drop('Species',axis=1))
```

```python
scaled_features
```

```python
new_dataset=pd.DataFrame(scaled_features,columns=dataset.columns[:-1])
```

```python
new_dataset
```

```python
from sklearn.model_selection import train_test_split
x=new_dataset
y=dataset['Species']
```

```python
 x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB, BernoulliNB, CategoricalNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
scores = []
```

```python
classifier = GaussianNB()
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test) 
scores.append(accuracy_score(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
cm 
```

```python
print(scores)
```

```python
classifier = BernoulliNB()
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test) 
scores.append(accuracy_score(y_test, y_pred))
bm= confusion_matrix(y_test, y_pred)
bm 
```
